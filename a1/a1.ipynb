```python
import os
import re
import pyecharts
import time

def ping():
    cmd = 'ping www.google.com -n 1'
    res = os.popen(cmd)
    output_str = res.read()   # 获得输出字符串
    #print(output_str)

    str1=output_str.split("\n")
    n=str1[2]
    #print(n)
    n=n.split(" ")
    time_str=n[4]
    time_str="时间=702ms" #时间=7ms test
    time_list=(re.findall(r"\d+", time_str))
    time_int=int(time_list[0])
    #print(time_int)  # 输出数字
    return(time_int)



from pyecharts.charts import Line
from pyecharts import options as opts

# 示例数据
cate = ["1"]
data1=[0]


    
num=ping()
data1 = data1+[num]
print(data1)
"""
折线图示例:
1. is_smooth 折线 OR 平滑
2. markline_opts 标记线 OR 标记点
"""
line = (Line()
    .add_xaxis(cate)
    .add_yaxis('电商渠道', [0], 
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average")]))
    .set_global_opts(title_opts=opts.TitleOpts(title="Line-基本示例", subtitle="我是副标题"))
    )




line.render_notebook()
time.sleep(1)
line = (Line()
    .add_xaxis(cate)
    .add_yaxis('电商渠道', [222], 
                markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="average")]))
    .set_global_opts(title_opts=opts.TitleOpts(title="Line-基本示例", subtitle="我是副标题"))
    )
```
